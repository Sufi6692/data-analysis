```python
# elif: grading:, A,B,C,...F
# if(condition1):
#   code
# elif(condition2):
#   code
# else:
#   code

marks= int(input("Enter your Marks:"))

if(marks>=85 and marks<=100):   #85-100
  print("Grade A")
elif(marks>=65 and marks<=84):  #65-84
  print("Grade B")
elif(marks>=45 and marks<=64): #45-64
  print("Grade C")
else:
  print("Grade F, Fail")


# c1: 78>=85 no
# c2: 78>=65 yes -> block will be executed

# c1: 35>=85 no
# c2: 35>=65 no
# c3: 35>=45 no
# if no condition sastisfy then go to else

# c1: 75>=85 no
# c2: 75>=45 yes
```

```python
#nested if: if statement inside a if statement

attendance=60
marks=67

if(attendance>=75):    # parent if
  if(marks>=85 and marks<=100):   # child if
    print("Grade A")
  elif(marks>=65 and marks<=84):
    print("Grade B")
  elif(marks>=45 and marks<=64):
    print("Grade C")
  else:
    print("Grade F, Fail")
else:
  print("No Grades can be provided as attendance is not above 75%")


# 91>=75 yes
#   67>=85 and 67<=100 no
#   67>=65 and 67<=84  yes
```

```python
l=["santosh","aarav","pardip","surya"]
for i in range(0,len(l)):
  print(i,"indexed value is",l[i])
```

```python
#reverse
l=["santosh","aarav","pardip","surya"]
for i in range(len(l)-1,-1,-1):
  print(i,"indexed value is",l[i])
```

```python
l=["santosh","aarav","pardip","surya"]
for i in l:
  print(i)
```

```python
#while: Runs till a conditions is true

#while(condition):
  #code

# 1-5

n=1
while(n<=5):    #1<=5
  print(n)
  n+=1
```

```python
#5-1
n=5
#if the while loop condition is false it will never go inside the loop
while(n>=1):
  print(n)
  n-=1
```

```python
#while
n=5
while(n==1):    #5==1 False
  print(n)
  n-=1
```

```python
#control Statements

# break, continue

#break: stop the loop forcefully
for i in range(1,11):
  if(i==6):
    break
  print(i)
```

```python
#continue: skip the current iteration

for i in range(1,11):
  if(i%2==1):
    continue
  print(i)
```

```python
#type 1: no return, no parameters

def sum():
  a=30
  b=50
  print(a+b)

sum()
```

```python
# type3: with return, no parameter
# return is used to alway return a value a function has been executed

def sum():
  a=45
  b=7
  return a+b

# total=sum()
# print(total)

print(sum())
```
