```python
# Check for Prime number take user input (loops and if statement)

a=int(input("enter the number"))
c=0
for i in range(1,a+1):        # 3   - 1-3
  if(a%i==0):
    c+=1

if(c==2):
  print("prime number")
else:
  print("not a prime number")
```

```python
# count even and odd number from a list

l=[65,30,89,45,54,10,18]
c=0  #even-count
c1=0 #odd-count
for i in range(0,len(l)):
  if(l[i]%2==0):
    c+=1
  else:
    c1+=1

print("even count:",c)
print("Odd count:",c1)
```

```python
lst= [-2, 65,32, 89, 0, 45, 54, 10, 18, 1, 21]
oddCounter = 0
evenCounter = 0
for i in range(0, len(lst)):
  if(lst[i] <= 0):
    continue
  else:
    if(lst[i] % 2 == 0):
      evenCounter += 1
    else:
      oddCounter += 1
print('Even numbers are', evenCounter)
print('Odd numbers are', oddCounter)
```
